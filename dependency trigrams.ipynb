```python
#data = pd.read_csv('dependency_parsed_english.csv')
data = pd.read_csv('dependency_parsed_chinese.csv')
data.head(20)
```

```python
trigrams = []
for index, row in data.iterrows():
    if row['head id'] < row['word id']:
        if row['head id']!=0:
            trigrams.append(str(row['deprel']) + ' ' + str(row['head']) + ' ' + str(row['word']))
        else:
            trigrams.append(str(row['deprel']) + ' ' + str(row['word']) + ' ' + str(row['word']))
    elif row['head id'] > row['word id']:
        if row['head id']!=0:
            trigrams.append(str(row['deprel']) + ' ' + str(row['word']) + ' ' + str(row['head']))
        else:
            trigrams.append(str(row['deprel']) + ' ' + str(row['word']) + ' ' + str(row['word']))
    #(label of word, head/governer of word, word)
    
```
