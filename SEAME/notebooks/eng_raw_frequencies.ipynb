```python
import urllib
import requests

raw_freqs = []

with open('first_cs_word.txt') as f:
    lines = f.readlines()
lines = [word[:-1] for word in lines]

for i in range(len(lines)):
    word = lines[i]
    if i%100==0:
        print(i)
    
    encoded_query = urllib.parse.quote(word)
    params = {'corpus': 'eng-us', 'query': encoded_query, 'topk': 1}
    params = '&'.join('{}={}'.format(name, value) for name, value in params.items())

    response = requests.get('https://api.phrasefinder.io/search?' + params)

    if response.status_code == 200:
        if response.json()['phrases'] == []:
            raw_freqs.append(0)
        else:
            raw_freqs.append(response.json()['phrases'][0]['mc'])
    else:
        raw_freqs.append('TODO')
```
