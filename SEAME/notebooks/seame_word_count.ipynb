```python
import pandas as pd
```

```python
df = pd.read_csv('seame_data_annotated.csv')
df.head()
```

```python
transcripts = list(df['filtered_transcript_2015'])
transcripts[:10]
```

```python
cleaned_transcripts = []

for transcript in transcripts:
    cleaned = transcript.strip('][').split(', ')
    final_cleaned = [word[1:-1] for word in cleaned if word.startswith("'") and word.endswith("'")]
    cleaned_transcripts.append(final_cleaned)
```

```python
total_count = 0

for transcript in cleaned_transcripts:
    total_count += len(transcript)
    
print(total_count)
```
