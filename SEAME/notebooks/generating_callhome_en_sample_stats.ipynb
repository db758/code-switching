```python
import spacy
import string
import pandas as pd
from collections import defaultdict
import json
```

```python
file = open('callhome_en_test_transcripts.txt', 'r')
lines = []
for line in file.readlines(): 
    lines.append(line[:-1])
```

```python
lines
```

```python
lines = [line.translate(str.maketrans('', '', string.punctuation))for line in lines]
```

```python
lines = [line.strip() for line in lines]
```

```python
lengths = []
for line in lines:
    for word in line.split():
        lengths.append(len(word))
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
tags = []
for line in lines:
    doc = nlp(line)
    #print(doc)
    for token in doc:
        #print(token)
        tags.append(token.tag_)
        #break
```

```python
coded_POS = []
for tag in tags:
    if tag == 'NN' or tag == 'NNP' or tag == 'NNS' or tag == 'NNPS':
        coded_POS.append(0)
    elif 'V' in tag:
        coded_POS.append(1)
    else:
        coded_POS.append(2)
```

```python
coded_POS = coded_POS[:len(lengths)]
```

```python
cols = {'lengths': lengths, 'POS': coded_POS} 
    
df = pd.DataFrame(cols)
df.head()
df.to_csv('callhome_en_lengths_POS.csv')
```

```python
total_count_en = 356033418959
unique_callhome_words = defaultdict()
for line in lines:
    for word in line.split():
        unique_callhome_words[word.lower()] = 1/total_count_en
```
