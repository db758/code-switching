```python
import csv
```

```python
lines = []
with open("merged_eng_callhome.txt") as tsv:
    for line in csv.reader(tsv, dialect="excel-tab"):
        lines.append(line)
```

```python
lines
```

```python
len(lines)
```

```python
lines = [line for line in lines if line[0]=='*A:' or line[0]=='*B:']
```

```python
len(lines)
```

```python
transcripts = [line[1:][0] for line in lines]
```

```python
transcripts = [line[:-15] for line in transcripts]
```

```python
cleaned_transcripts = []
```

```python
for line in transcripts:
    clean = []
    for word in line.split():
        if '&=' not in word:
            clean.append(word)
    cleaned_transcripts.append(" ".join(clean))
```

```python
cleaned_transcripts = [line for line in cleaned_transcripts if line != '.']
```

```python
len(cleaned_transcripts)
```

```python
0.8*len(cleaned_transcripts)
```

```python
train_transcripts = cleaned_transcripts[:28369]
test_transcripts = cleaned_transcripts[28369:]
```

```python
len(train_transcripts)
```

```python
len(test_transcripts)
```

```python
file = open('callhome_en_train_transcripts.txt','w')
for line in train_transcripts:
    file.write(line+"\n")
file.close()
```

```python
file = open('callhome_en_test_transcripts.txt','w')
for line in test_transcripts:
    file.write(line+"\n")
file.close()
```

```python
file = open('merged_hub5.txt', 'r')
lines = []
for line in file.readlines(): 
    lines.append(line.split())
```

```python
lines = [line for line in lines if len(line)>0]
```

```python
lines = [line for line in lines if line[0] != '#Language:' and line[0] != '#File']
```

```python
lines = [" ".join(line[3:]) for line in lines]
```

```python
lines
```

```python
cleaned_lines = []
for line in lines:
    clean = []
    for word in line.split():
        if '%' not in word and '[' not in word and '<' not in word and '{' not in word and not word.isascii():
            clean.append(word)
    cleaned_lines.append(" ".join(clean))
```

```python
cleaned_lines = [line for line in cleaned_lines if len(line)>0]
```

```python
cleaned_lines
```

```python
len(cleaned_lines)
```

```python
0.8*len(cleaned_lines)
```

```python
train_transcripts = cleaned_lines[:25320]
test_transcripts = cleaned_lines[25320:]
```

```python
len(train_transcripts)
```

```python
len(test_transcripts)
```

```python
file = open('hub5_train_transcripts.txt','w')
for line in train_transcripts:
    file.write(line+"\n")
file.close()
```

```python
file = open('hub5_test_transcripts.txt','w')
for line in test_transcripts:
    file.write(line+"\n")
file.close()
```
