```python
new_df = pd.DataFrame(columns = ['university', 'sent_id', 'original_sentence', 'translation', 'word_id', 'first_cs_word_form', 'first_cs_word_translation', 'governer_first_cs_word', 'preceding_word'])
```

```python
new_df
```

```python
prec = []

for index, row in df.iterrows():
    if row['sent_type'] == 'code-switch':
        og_list_of_words = list(list(df[index:index+1]['original_sentence'])[0].split())
        chinese_list_of_words = list(list(df[index:index+1]['zh_translation'])[0].split())
        #print(og_list_of_words)
        #print()
        #print(chinese_list_of_words)
        #print(row['sent_id'])
        try:
        
            if row['cs_word_id'] == 1:
                prec.append('</s>')
    
            else: #word_id!=1 and it is a root 
                prec.append(og_list_of_words[row['cs_word_id']-2])
                
        except: 
            pass
            
```

```python
gov = []

for index, row in df.iterrows():
    og_list_of_words = list(list(df[index:index+1]['original_sentence'])[0].split())
    #print(og_list_of_words)
    if row['if_it_is_root'] == 1:
        try:
            word_index = row["cs_word_id"]
            #print(word_index)
            word = og_list_of_words[word_index-1]
        except:
            word = 'NaN'
    else:
        try:
            word_index = row["governor_first_cs_word_trans"]
            word = og_list_of_words[word_index-1]
            #print(word_index)
        except:
            word = 'NaN'
    gov.append(word)

            
```

```python
print(gov.count('NaN'))
```

```python
textfile = open("redone_cs_governor_words.txt", "w")
for element in gov:
    textfile.write(element + "\n")
textfile.close()
```

```python
textfile = open("redone_cs_preceding_words.txt", "w")
for element in prec:
    textfile.write(element + "\n")
textfile.close()
```

```python
for index, row in df.iterrows():
    if row['sent_type'] == 'code-switch':
        og_list_of_words = list(list(df[index:index+1]['original_sentence'])[0].split())
        chinese_list_of_words = list(list(df[index:index+1]['zh_translation'])[0].split())
        #print(og_list_of_words)
        #print()
        #print(chinese_list_of_words)
        #print(row['sent_id'])
        try:
        
            if row['word_id'] == 1 and row['if_it_is_root'] == 0:

                new_df = new_df.append({'university': row['university'], 'sent_id': row['sent_id'], 
                                   'original_sentence': row['original_sentence'], 'translation': row['translation'],
                                   'word_id': row['word_id'], 'first_cs_word_form': row['first_cs_word_form'],
                                   'first_cs_word_translation': row['first_cs_word_translation'],
                                   'governer_first_cs_word': og_list_of_words[row['governor_first_cs_word_trans']-1], 
                                        'preceding_word': 'none' }, ignore_index=True)

            elif row['word_id'] == 1 and row['if_it_is_root'] == 1:
                new_df = new_df.append({'university': row['university'], 'sent_id': row['sent_id'], 
                                   'original_sentence': row['original_sentence'], 'translation': row['translation'],
                                   'word_id': row['word_id'], 'first_cs_word_form': row['first_cs_word_form'],
                                   'first_cs_word_translation': row['first_cs_word_translation'],
                                   'governer_first_cs_word': row['first_cs_row_translation'], 
                                        'preceding_word': 'none' }, ignore_index=True)

            elif row['word_id'] != 1 and row['if_it_is_root'] == 0:
                new_df = new_df.append({'university': row['university'], 'sent_id': row['sent_id'], 
                                   'original_sentence': row['original_sentence'], 'translation': row['translation'],
                                   'word_id': row['word_id'], 'first_cs_word_form': row['first_cs_word_form'],
                                   'first_cs_word_translation': row['first_cs_word_translation'],
                                   'governer_first_cs_word': og_list_of_words[row['governor_first_cs_word_trans']-1], 
                                        'preceding_word': og_list_of_words[row['word_id']-2] }, ignore_index=True)

            else: #word_id!=1 and it is a root 
                new_df = new_df.append({'university': row['university'], 'sent_id': row['sent_id'], 
                                   'original_sentence': row['original_sentence'], 'translation': row['translation'],
                                   'word_id': row['word_id'], 'first_cs_word_form': row['first_cs_word_form'],
                                   'first_cs_word_translation': row['first_cs_word_translation'],
                                   'governer_first_cs_word': row['first_cs_word_translation'], 
                                        'preceding_word': og_list_of_words[row['word_id']-2] }, ignore_index=True)
        except: 
            pass
            
```
