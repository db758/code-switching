```python
import pandas as pd
```

```python
df = pd.read_csv('redoing_cs_data.csv')
df.head()
```

```python
sents = df[['zh_translation', 'eng_translation', 'first_cs_word_form']]
```

```python
sents.head()
```

```python
pairs = []
for index, row in sents.iterrows():
    tup = (row['zh_translation'], row['eng_translation'], row['first_cs_word_form'])
    pairs.append(tup)
```

```python
import random
```

```python
rand = random.choices(pairs, k=148)
```

```python
zh_translation = [pair[0] for pair in rand]
eng_translation = [pair[1] for pair in rand]
cs_word = [pair[2] for pair in rand]
```

```python
final = pd.DataFrame(list(zip(zh_translation, eng_translation, cs_word)),
               columns =['zh_translation', 'eng_translation', 'first_cs_word_form'])
final
```

```python
final.to_csv('for_precision_checking.csv')
```
