```python
import pandas as pd
import numpy as np
df = pd.read_csv('redoing_cs_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
with open('cs2_indices.txt') as f:
    cs2_lines = f.readlines()
```

```python
cs2_lines = [item[:-1] for item in cs2_lines]
cs2_lines
```

```python
with open('cs1_indices.txt') as f:
    cs1_lines = f.readlines()
```

```python
cs1_lines = [item[:-1] for item in cs1_lines]
cs1_lines
```

```python
df = df[['original_sentence', 'zh_word_index', 'cs_back_word']]
df['cs1_index'] = cs1_lines
df.head()
```

```python
df = df[['cs1_index', 'cs_back_word', 'original_sentence']]
```

```python
len(df)
```

```python
df = df.dropna()
len(df)
```

```python
df['cs2_index'] = cs2_lines
df.head()
```

```python
df['distance'] = df.apply(lambda row: int(row.cs2_index) - int(row.cs1_index), axis=1)
df.head()
```

```python
df = df[['original_sentence', 'cs1_index', 'cs2_index', 'distance']]
df.head()
```

```python
df.to_csv('cs2-cs1 distances.csv')
```
