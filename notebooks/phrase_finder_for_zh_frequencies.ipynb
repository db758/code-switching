```python
file = open('code_switched_sentences_cs_back.txt', 'r')
lines = []
for line in file.readlines(): 
    lines.append(line[:-1])
```

```python
lines
```

```python
match_counts = []

for char in lines:
    encoded_query = urllib.parse.quote(str(char))
    params = {'corpus': 'chi', 'query': encoded_query, 'topk': 3}
    params = '&'.join('{}={}'.format(name, value) for name, value in params.items())

    response = requests.get('https://api.phrasefinder.io/search?' + params)

    assert response.status_code == 200

    try:
        match_counts.append(dict(response.json())['phrases'][0]['mc'])
    except:
        match_counts.append(1)
```

```python
match_counts
```

```python
import csv
```

```python
sum = 0
with open("googlebooks-chi-sim-all-totalcounts-20120701.txt") as tsv:
    for line in csv.reader(tsv, dialect="excel-tab"): 
        full_list = line
        #print(full_list)
        for elt in full_list:
            if elt != ' ' and elt != '':
                #print(elt)
                #print(str.split(elt, ',')[1])
                sum += int(str.split(elt, ',')[1])
print(sum)
```

```python
import numpy as np
```

```python
frequencies = [-np.log(num/sum) for num in match_counts]
```

```python
len(frequencies) == len(match_counts)
```

```python
import spacy
from nltk import tokenize
nlp = spacy.load("zh_core_web_sm")
```

```python
file = open('code_switched_sentences_cs_back.txt', 'r')
p = file.read()
tokenized = p.splitlines()
```

```python
tokenized
```

```python
tags = []
for word in tokenized:
    doc = nlp(word)
    for token in doc:
        tags.append(token.tag_)
        break
```

```python
tags
```

```python
file = open('cs_sentences_cs_back_surp.txt', 'r')
surps = []
for line in file.readlines(): 
    surps.append(line[:-1])
```

```python
surps
```

```python
len(frequencies) == len(surps) == len(lengths) == len(tags)
```

```python
import pandas as pd
```

```python
all_rows = []
for i in range(len(surps)):
    row = []
    row.append(frequencies[i])
    row.append(tags[i])
    row.append(lengths[i])
    row.append(surps[i])
    all_rows.append(row)
```

```python
len(all_rows)
```

```python
df = pd.DataFrame (all_rows, columns = ['word_freq_zh', 'cs_word_pos_zh', 'cs_word_length_zh', 'ngram_surp_zh'])
```
