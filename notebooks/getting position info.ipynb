```python
import pandas as pd
random_non_cs = pd.read_csv("new_random_non_cs_1_data.csv")
non_cs = pd.read_csv("non_cs_sentences_full.csv")
```

```python
random_non_cs.head()
```

```python
words = list(random_non_cs['random_non_cs_1_word'])
```

```python
sentences = list(non_cs['sentence'])
```

```python
len(words) == len(sentences)
```

```python
sentences[0]
```

```python
indices = []
for i in range(len(words)):
    word = words[i]
    sentence = sentences[i]
    split = sentence.split()
    indices.append(split.index(word))
```

```python
indices
```

```python
random_non_cs['zh_word_index'] = indices
```

```python
random_non_cs.head()
```

```python
random_non_cs.to_csv("new_random_non_cs_1_data_with_indices.csv")
```
