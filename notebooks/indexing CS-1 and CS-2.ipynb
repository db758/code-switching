```python
import pandas as pd
```

```python
cs_data = pd.read_csv('redoing_cs_data.csv')
cs_data.head()
```

```python
cs_data.columns
```

```python
sentences = list(cs_data['zh_translation'])
cs1_words = list(cs_data['first_cs_word_translation'])
cs2_words = list(cs_data['cs_back_word'])
```

```python
sentences[0].split()
```

```python
sentences = [sentence.split() for sentence in sentences]
```

```python
cs2_words
```

```python
cs1_indices = []
for i in range(len(sentences)):
    cs1_indices.append(sentences[i].index(cs1_words[i]))
```

```python
cs1_indices
```

```python
cs2_indices = []
for i in range(len(sentences)):
    try:
        if type(cs2_words[i]) is str:
            cs2_indices.append(sentences[i].index(cs2_words[i]))
    except:
        cs2_indices.append(len(sentences[i])) #for words that aren't in the chinese translation
    # if the cs2 word is nan, no need to index it
```

```python
textfile = open("cs2_indices.txt", "w")
for element in cs2_indices:
    textfile.write(str(element) + "\n")
textfile.close()
```

```python
textfile = open("cs1_indices.txt", "w")
for element in cs1_indices:
    textfile.write(str(element) + "\n")
textfile.close()
```
