```python
import pandas as pd
df = pd.read_csv("redoing_cs_data.csv")
df.head()
```

```python
df.columns
```

```python
df = df[['original_sentence', 'cs_back_word']]
df.head()
```

```python
sentences = []
for index, row in df.iterrows():
    if type(row['cs_back_word']) is float:
        sentences.append(row['original_sentence'])
len(sentences)
```

```python
split_sentences = []
for sent in sentences:
    split_sentences.append(sent.split())
```

```python
split_sentences
```

```python
counts = []
for sent in split_sentences:
    total = 0
    for word in sent:
        if word.isascii():
            total +=1
    counts.append(total)
```

```python
lengths = [len(sent) for sent in split_sentences]
```

```python
df_new = pd.DataFrame(list(zip(sentences, counts, lengths)),
               columns =['sents', 'counts', 'lengths'])
```
