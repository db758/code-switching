```python
import pandas as pd
import csv 
import stanza
from nltk import tokenize

nlp = stanza.Pipeline(lang='en', processors='tokenize,pos,lemma,depparse')
```

```python
data = pd.read_csv("non_cs_sentences_updated.csv")
data.head(10)
```

```python
sentences = []
for index, row in data.iterrows():
    sentences.append(row['eng_translation'])
```

```python
result = ['NA'] * len(sentences)
for i in range(len(sentences)):
    doc = nlp(sentences[i])
    for sent in doc.sentences:
        for word in sent.words:
            if word.text.lower() == data['eng_first_cs_word_translation'][i]:
                print('word: '+word.text)
                print('head: '+sent.words[word.head-1].text if word.head > 0 else "root")
                if word.id > word.head:
                    print(word.deprel + " " + sent.words[word.head-1].text + " " + word.text)
                    result[i] = word.deprel + " " + sent.words[word.head-1].text + " " + word.text
                elif word.id < word.head:
                    print(word.deprel + " " + word.text + " " + sent.words[word.head-1].text)
                    result[i] = word.deprel + " " + word.text + " " + sent.words[word.head-1].text
                else: 
                    print(word.deprel + " " + word.text)
                    result[i] = word.deprel + " " + word.text
    #print(*[f'id: {word.id}\tword: {word.text}\thead id: {word.head}\thead: {sent.words[word.head-1].text if word.head > 0 else "root"}\tdeprel: {word.deprel}' for sent in doc.sentences for word in sent.words], sep='\n')

                 
                 
                 
```

```python
new_data = pd.read_csv("non_cs_sentences.csv")
new_data.head()
```

```python
all_list = []
print(all_list)
for index, row in data.iterrows():
    sent = data['translation'][index]
    segmented = jieba.lcut(sent)
    cleaned_seg = [elt for elt in segmented if ' ' not in elt]
    word = cleaned_seg[data['word_id'][index]-1]
    head_id = new_data['governor_first_cs_word_trans'][index]
    head = cleaned_seg[head_id-1]
    deprel = new_data['deprel_first_cs_word_trans'][index]
    if  data['word_id'][index] < head_id:
        all_list.append(deprel+" "+word+" "+head)
    elif data['word_id'][index] > head_id:
        all_list.append(deprel+" "+head+" "+word)
    else:
        all_list.append(deprel+" "+word)
```
